```python
Data Cleaning


Q1. Check whether the dataset contains any null or missing values. If found, handle
them appropriately.
Q2. Identify and remove any duplicate rows from the dataset.
Q3. Check the 'age' column for unrealistic values (e.g., 0 or 200). Filter and display such
rows if any exist.
Q4. The 'loyalty_score' should be between 0 and 10. Check if any values fall outside this
range.
Q5. Verify that all numerical columns have the correct data type. Convert any column
that has an incorrect type.
Q6. Detect outliers in the 'purchase_amount' and 'annual_income' columns using the
IQR method.
```

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r"C:\Users\Parth\OneDrive\Desktop\student_prrform_analyzer\Customer Purchasing Behaviors.csv")
print(df.head())
print(df.shape)
df
```

```python
# Q1. Check whether the dataset contains any null or missing values. If found, handle them appropriately.

print(df.isnull().sum())
print("Total missing Values:", df.isnull().sum().sum())

#handle if any
df['age'].fillna(df['age'].median(),inplace=True)
df['annual_income'].fillna(df['annual_income'].median(),inplace=True)
df['purchase_amount'].fillna(df['purchase_amount'].median(),inplace=True)
df['loyalty_score'].fillna(df['loyalty_score'].median(),inplace=True)
df['region'].fillna(df['region'].mode()[0],inplace=True)
df['purchase_frequency'].fillna(df['purchase_frequency'].median(),inplace=True)

print("After handling: ", df.isnull().sum().sum())
```

```python
#Q2. Identify and remove any duplicate rows from the dataset.

# check

Duplicate_rows= df.duplicated().sum()
print("Duplicate rows:",Duplicate_rows)

# remove

df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
#Q3. Check the 'age' column for unrealistic values (e.g., 0 or 200). Filter and display such rows if any exist.

# filter rows with unrealstic age
age = df[(df['age'] <=0) | (df['age'] >= 120)]
print("Unrealistic age: ",len(age))

# Remove
df= df[(df['age'] > 0) & (df['age'] <= 120)]
print("Shape after cleaning age: ",df.shape)
```

```python
#Q4. The 'loyalty_score' should be between 0 and 10. Check if any values fall outside this range.

#check

terrible_score= df[(df['loyalty_score'] < 0) | (df['loyalty_score'] > 10)]
print("Invalid loyalty_score: ",len(terrible_score))

#Remove
df= df[(df['loyalty_score'] >= 0) & (df['loyalty_score'] <= 10)]
print("shape after cleaning loyalty_score: ",df.shape)
```

```python
#Q5. Verify that all numerical columns have the correct data type. Convert any column that has an incorrect type.

print(df.dtypes)

# convert
df['age']= pd.to_numeric(df['age'])
df['annual_income']= pd.to_numeric(df['annual_income'])
df['purchase_amount']= pd.to_numeric(df['purchase_amount'])
df['loyalty_score']= df['loyalty_score'].astype(float)
df['purchase_frequency']= pd.to_numeric(df['purchase_frequency'])

# verify
print("shape after conversion: ",df.dtypes)
```

```python
#Q6. Detect outliers in the 'purchase_amount' and 'annual_income' columns using the IQR method.

for col in ['annual_income','purchase_amount']:
    Q1= df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)

    IQR= Q3 - Q1
    lower= Q1 - 1.5*IQR
    upper= Q3 + 1.5*IQR

    outliers= df[(df[col] < lower) | (df[col] > upper)]

    print(f"\n---------- {col} ----------")
    print(f"Q1= {Q1}, Q3= {Q3}, IQR= {IQR}")
    print(f"Lower Bound= {lower}, Upper Bound= {upper}")
    print(f"Outliers Found: {len(outliers)}")
    print(outliers[[col]])
```

```python
GroupBy & Aggregation
Q7. Find the average purchase amount for each region.
Q8. Count the number of customers in each region.
Q9. Divide customers into age groups (20-29, 30-39, 40-49, 50+) and find the average
loyalty score for each group.
Q10. Which region has the highest average purchase frequency?
Q11. Find the maximum and minimum annual income for each region.
Q12. Divide customers into three loyalty categories — Low (0–4), Medium (4–7), High
(7–10) — and count customers in each category.
Q13. Find the total purchase amount for each age group.
Q14. Which region has the highest average loyalty score?
Q15. Divide customers into two income groups — Low Income (< 50,000) and High
Income (>= 50,000) — and compare their average purchase amount.
```

```python
#Q7. Find the average purchase amount for each region.

q7= df.groupby('region')['purchase_amount'].mean()
q7
```

```python
#Q8. Count the number of customers in each region.

q8=df.groupby('region')['user_id'].count()
q8
```

```python
#Q9. Divide customers into age groups (20-29, 30-39, 40-49, 50+) and find the average loyalty score for each group.

# create age group
bins= [20,30,40,50,100]
labels= ['20-29','30-39','40-49','50+']

df['age_group']= pd.cut(df['age'], bins=bins, labels=labels, right=True)

#groupby
q9= df.groupby('age_group')['loyalty_score'].mean()
q9
```

```python
#Q10. Which region has the highest average purchase frequency? 

q10= df.groupby('region')['purchase_frequency'].mean()
q10
```

```python
#Q11. Find the maximum and minimum annual income for each region.

q11= df.groupby('region')['annual_income'].agg(['max','min'])
q11
```

```python
#Q12. Divide customers into three loyalty categories — Low (0–4), Medium (4–7), High (7–10) — and count customers in each category.

#create bins
bins= [0,4,7,10]
labels= ["Low", "Medium", "High"]

df['loyalty_categories']= pd.cut(df['loyalty_score'], bins=bins, labels=labels, right=True)

#Groupby
q12= df.groupby('loyalty_categories')['user_id'].count()
q12
```

```python
#Q13. Find the total purchase amount for each age group.

q13= df.groupby('age_group')['purchase_amount'].sum()
q13
```

```python
#Q14. Which region has the highest average loyalty score?

q14= df.groupby('region')['loyalty_score'].mean()
print(q14)
print("\nHighest:", q14.idxmax(), "→", q14.max())
```

```python
#Q15. Divide customers into two income groups — Low Income (< 50,000) and High Income (>= 50,000) — and compare their average purchase amount.

#create bins
bins= [0,50000,float('inf')]
labels= ["Low Income","High Income"]

df['income_group']= pd.cut(df['annual_income'], bins=bins, labels=labels, right=False)
df['income_group']

#groupby
q15= df.groupby('income_group')['purchase_amount'].mean()
q15
```

```python
Pivot Table
Q1.Region-wise average purchase amount pivot banao
Q2.Age group vs Region — average loyalty score
Q3.Income group vs Region — total purchase amount
Q4.Loyalty category vs Region — customer count
Q5.Grand total ke saath region-wise summary
```

```python
#Q1.Region-wise average purchase amount pivot banao

p1 = pd.pivot_table(df,
                    index=["region"],
                    values=['purchase_amount'],
                    aggfunc="mean").round(2)
p1
```

```python
#Q2.Age group vs Region — average loyalty score

p2= pd.pivot_table(df,
                  index=["age_group"],
                  columns=['region'],
                  values=['loyalty_score'],
                  aggfunc="mean").round(2)
p2
```

```python
#Q3.Income group vs Region — total purchase amount

p3= pd.pivot_table(df,
                  index=["income_group"],
                   columns=['region'],
                   values=['purchase_amount'],
                   aggfunc="sum")
p3
```

```python
#Q4.Loyalty category vs Region — customer count

p4= pd.pivot_table(df,
                  index=["loyalty_categories"],
                  columns=["region"],
                  values=["user_id"],
                  aggfunc="count")
p4
```

```python
#Q5.Grand total ke saath region-wise summary

p5= pd.pivot_table(df,
                  index=["region"],
                  values=["purchase_amount","loyalty_score"],
                  aggfunc="mean").round(2)
p5
```
